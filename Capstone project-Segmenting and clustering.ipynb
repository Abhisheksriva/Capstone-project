In this project we have to scraping the Wikipedia page for data of the postal codes of Canada.
After that for clustering we need to process and clean data. Then use K Means for clusterning and then clusters are plotted using the Folium Library.
The Boroughs containing the name 'Toronto' in it are first plotted and then clustered and plotted again.

```python
#Import libraries
import numpy as np # library to handle data in a vectorized manner

import pandas as pd # library for data analsysis
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
!pip install beautifulsoup4
!pip install lxml
import random # library for random number generation

!conda install -c conda-forge geopy --yes 
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors
from bs4 import BeautifulSoup

# import k-means from clustering stage
from sklearn.cluster import KMeans

!conda install -c conda-forge folium=0.5.0 --yes 
import folium # map rendering library

print('Libraries imported.')
```

# Scraping the Wikipedia page for the table of postal codes of Canada

```python
data=requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(data,'lxml')
from IPython.display import display_html
table=str(soup.table)
display_html(table,raw=True)

```

# Convert html table into pandas dataframe

```python
table_contents=[]
table=soup.find('table')
for row in table.findAll('td'):
    cell = {}
    if row.span.text=='Not assigned':
        pass
    else:
        cell['Postal Code'] = row.p.text[:3]
        cell['Borough'] = (row.span.text).split('(')[0]
        cell['Neighborhood'] = (((((row.span.text).split('(')[1]).strip(')')).replace(' /',',')).replace(')',' ')).strip(' ')
        table_contents.append(cell)
table_contents        
```

```python
df=pd.DataFrame(table_contents)
df['Borough']=df['Borough'].replace({'Downtown TorontoStn A PO Boxes25 The Esplanade':'Downtown Toronto Stn A',
                                             'East TorontoBusiness reply mail Processing Centre969 Eastern':'East Toronto Business',
                                             'EtobicokeNorthwest':'Etobicoke Northwest','East YorkEast Toronto':'East York/East Toronto',
                                             'MississaugaCanada Post Gateway Processing Centre':'Mississauga'})
df
```

# Importing the csv file conatining the latitudes and longitudes for various neighbourhoods in Canada

```python
# To utilize the Foursquare location data, we need to get the latitude and the longitude coordinates of each neighborhood
lati_longi=pd.read_csv('Geospatial_Coordinates(1).csv')
lati_longi.head()
```

```python
#To get latitude and longitude of all neighborhoods in Canada,concat table df and lati_longi table
df.rename(columns={'Postal Code':'Postcode'},inplace=True)
df1 = pd.merge(df,lati_longi,on='Postcode')
df1.head(11)
```

# Explore and cluster the neighborhoods in the city of Toronto

**Clustering and the plotting of the neighbourhoods of Canada which contain Toronto in their Borough**

```python
#Extract all rows which contain Toronto word
df2=df1[df1['Borough'].str.contains('Toronto',regex=False)]
df2
```

```python
map_Toronto = folium.Map(location=[43.6534817, -79.3839347], zoom_start=11)

# add markers to map
for lat, lng, borough,neighborhood in zip(df2['Latitude'],df2['Longitude'],df2['Borough'], df2['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_Toronto)  
    
map_Toronto
```

# Cluster Neighborhoods

 **Use k-means to cluster the neighborhood into 5 clusters**

```python
# set number of clusters
k=5
toronto_clustering = df2.drop(['Postcode','Borough','Neighborhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
df2.insert(0, 'Cluster Labels', kmeans.labels_)
df2
```

**let's visualize the resulting clusters**

```python
# create map
map_clusters = folium.Map(location=[43.6534817, -79.3839347], zoom_start=11)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(df2['Latitude'], df2['Longitude'], df2['Neighborhood'],df2['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
